```python
#12 Following were the Most Occurence of the People In the Dataset

states = ['Thomas', 'Watson', 'Kelly', 'Farrell', 'Tucker', 'Higgins', 'Parker', 'Hall']
 
val = [17,16,15,15,15,15,15,14]
wp = { 'linewidth' : 2, 'edgecolor' : "blue" }

colors = ( "beige", "cyan", "orange", "pink", "red", "black", "green", "yellow")
# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(val, labels = states, shadow = True, startangle = 90, wedgeprops = wp, textprops = dict(color ="blue"),colors = colors)

# show plot
plt.show()
```
